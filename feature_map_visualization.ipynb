Feature map visualization
OmniXAI provides a simple explainer for visualizing feature maps of a specified layer.

```python
import torch
from torchvision import transforms
from PIL import Image as PilImage
from omnixai.data.image import Image
from omnixai.explainers.vision.specific.feature_visualization.visualizer import FeatureMapVisualizer
```

```python
img =  Image(PilImage.open(r"images_resize0.5_contrastReduce21-40_num_image10000\part_whole_test\0\s_plus+1.png").convert('RGB'))
```

```python
img_size = 224
model = torch.load(r'models\resnet18_is224_bs4_e10_i10000_resize0.5_contrastReduce1-20_num_image10000.pth')
transform = transforms.Compose([
        transforms.Resize(size=img_size),
        transforms.ToTensor(),
        ])
preprocess = lambda ims: torch.stack([transform(im.to_pil()) for im in ims])
```

```python
explainer = FeatureMapVisualizer(
    model=model,
    target_layer=model.layer4[-1],
    preprocess_function=preprocess
)
explanations = explainer.explain(img)
explanations.ipython_plot()
```
